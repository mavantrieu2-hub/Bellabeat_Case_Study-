3. Perform calculations
</H2>

> **Activity Minutes**:

- **Sedentary Minutes**: Count minutes when intensity by minute equal 0.
- **Light Active Minutes**: Count minutes when intensity by minute equal 1.
- **Moderate Active Minutes**: Count minutes when intensity minute equal 2.
- **Very Active Miniutes**: Count minutes when intensity minute equal 3.
- **Total Active Minutes**: Total of Light Active Minutes, Moderate Active Minutes, Very Active Minutes.
- **Purpose**: Evaluation periods active or sedentary by day or hour

> **Rate**:

- **Total sedentary**: Total sedentary minutes at all periods of all users.
- **Total Minutes**: Total of Sedentary active minutes, Light active minutes, Moderate active minutes, Very active minutes.
- **Sedentary Rate**: Total sedentary divide Total minutes.

- **Purpose**: Evaluation relationship between Sedentary rate with orther metrics.

> **Key metrics per daily**:

- **Median Total Active Minutes**: The middle number of minutes in each day of the week when arranged in order.
- **Purpose**: Evaluation behavioral trends between day of week.  

> **Key metrics per hour**:

- **Average Total Active Minutes**: Total active minutes divide number of values.

- **Purpose**: Evaluation behavioral trends between hour of day.
